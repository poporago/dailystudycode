```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)

ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)

# equation ,   t .   t 3 
t=np.linspace(0 ,2, 101)
x=np.sin(6*t)
y=1/4*t
z=t**2/2

ax.plot3D(x,y,z, c='k')
ax.plot([x[0]],[y[0]],[z[0]], 'o', markersize=10, color='k', label="t= {:.2f}".format(t[0]))
ax.plot([x[50]],[y[50]],[z[50]], '^', markersize=10, color='k', label="t= {:.2f}".format(t[50]))
ax.plot([x[-1]],[y[-1]],[z[-1]], '*', markersize=10, color='k', label="t= {:.2f}".format(t[-1]))

ax.legend(fontsize=15, loc='upper right')

plt.show()
# (2) (t) 
```

```python
plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)

ax.set_xlabel(r'$x_1$', fontsize=20)
ax.set_ylabel(r'$x_2$', fontsize=20)
ax.set_zlabel(r'$z$', fontsize=20)

X1= np.linspace(-2, 2, 50)
X2= np.linspace(-1, 3, 50)
Z=50*(X2-X1**2)**2 + (2-X1)**2

ax.scatter3D(X1,X2,Z, marker='*')

plt.show()
```

```python
from sklearn.datasets import load_digits
from sklearn.manifold import MDS
import matplotlib.pyplot as plt

# Load digits dataset (8x8 images of digits).
digits = load_digits()

# Apply MDS to the data.
mds = MDS(n_components=2)
mds_result = mds.fit_transform(digits.data)

# Plot the result.
plt.scatter(mds_result[:, 0], mds_result[:, 1], c=digits.target)
plt.colorbar()
plt.show()1
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Make data.
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the surface with color mapping.
plt.imshow(Z, cmap='hot', interpolation='nearest')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set up a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Generate X, Y coordinates
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Generate multiple Z coordinates for different hyperbolic functions
Z1 = np.sqrt(X**2 - Y**2)  # Hyperbolic function 1
Z2 = np.cosh(X) - np.cosh(Y)  # Hyperbolic function 2
Z3 = np.sinh(X) - np.sinh(Y)  # Hyperbolic function 3

# Plot each hyperbolic function with different colors
ax.plot_surface(X, Y, Z1, color='blue', alpha=0.5, rstride=100, cstride=100)
ax.plot_surface(X, Y, Z2, color='yellow', alpha=0.5, rstride=100, cstride=100)
ax.plot_surface(X, Y, Z3, color='cyan', alpha=0.5, facecolors='r', rstride=100, cstride=100, shade=False)

# Set plot view angle for better visibility
ax.view_init(elev=20, azim=-35)
plt.show()
```
